```python
# Question 3

def leapYear(year):
    if (year % 400) == 0:
        return True
    elif (year % 100) == 0:
        if ((year / 100) % 4) == 0:
            return True
        else:
            return False
    elif (year % 4) == 0:
        return True
    else:
        return False

def calender(startDay, year):
    days = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']
    months = ['Jan', 'Feb', 'March', 'April', 'May', 'June', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    
    leap = leapYear(year)
    x = days.index(startDay)
    y = x
    
    for i in months:
        print('\n' + i + '\n')
        
        if i == 'Jan' or i == 'March' or i == 'May' or i == 'July' or i == 'Aug' or i == 'Oct' or i == 'Dec':
            for j in range(0, 31):
                print('Date: - ', j+1, ' Day: - ', days[((j % 7) + x) % 7])
                y = ((j % 7) + x) % 7
            else:
                x = y + 1
        elif i == 'April' or i == 'June' or i == 'Sep' or i == 'Nov':
            for j in range(0, 30):
                print('Date: - ', j+1, ' Day: - ', days[((j % 7) + x) % 7])
                y = ((j % 7) + x) % 7
            else:
                x = y + 1
        else:
            if leap == True:
                for j in range(0, 29):
                    print('Date: - ', j+1, ' Day: - ', days[((j % 7) + x) % 7])
                    y = ((j % 7) + x) % 7
                else:
                    x = y + 1
            else:
                for j in range(0, 28):
                    print('Date: - ', j+1, ' Day: - ', days[((j % 7) + x) % 7])
                    y = ((j % 7) + x) % 7
                else:
                    x = y + 1
                    
                    
calender('Fri', 2021) # Example of 2021 calender which started with Friday
```

```python
# Question 4

def solveQuadratic(a, b, c): # assuming quadratic equation as a*X^2 + b*X + c
    
    d = b**2 - 4*a*c
    
    if d == 0:
        print("Only one solution: - ", int((-b + d**(1/2)) / (2*a)))
    elif d > 0:
        print("Solution 1: - ", int((-b + d**(1/2)) / (2*a)))
        print("Solution 2: - ", int((-b - d**(1/2)) / (2*a)))
    else:
        print("Sorry no solutions!!!")
        
solveQuadratic(1, -4, 4) # Solution of X^2 - 4X + 4 which is (X - 2)^2
```
