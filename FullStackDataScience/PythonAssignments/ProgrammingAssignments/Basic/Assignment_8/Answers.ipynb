```python
# Question 1

def addMatrices():
    try:
        m = int(input("Enter number of rows: - "))
        n = int(input("Enter numbers of columns: - "))
        
        a = []
        b = []
        c = []
        
        print("\nElements of matix 1: - ")
        for i in range(m):
            x = []
            for j in range(n):
                x.append(float(input(f"Enter element {i+1} {j+1}: - ")))
            a.append(x)
            
        print("\nElements of matrix 2: - ")
        for i in range(m):
            x = []
            for j in range(n):
                x.append(float(input(f"Enter element {i+1} {j+1}: - ")))
            b.append(x)
            
        for i in range(m):
            x = []
            for j in range(n):
                x.append(a[i][j] + b[i][j])
            c.append(x)
            
        print("\nAfter adding two matrices: -\n")
        for i in range(m):
            print(c[i])
        
    except Exception as e:
        print(e)
        
addMatrices()
```

```python
# Question 2

def multiplyMatrices():
    try:
        m1 = int(input("Enter number of rows of matrix 1: - "))
        n1 = int(input("Enter numbers of columns of matrix 1: - "))
        
        m2 = int(input("Enter number of rows of matrix 2: - "))
        n2 = int(input("Enter numbers of columns of matrix 2: - "))
        
        if n1 != m2:
            raise Exception("\nNo. of columns in matrix 1 is not equal to No. pf rows in matrix 2!!!\nCan't multiply matrices with given dimensions")
        
        a = []
        b = []
        c = []
        
        print("\nElements of matix 1: - ")
        for i in range(m1):
            x = []
            for j in range(n1):
                x.append(float(input(f"Enter element {i+1} {j+1}: - ")))
            a.append(x)
            
        print("\nElements of matrix 2: - ")
        for i in range(m2):
            x = []
            for j in range(n2):
                x.append(float(input(f"Enter element {i+1} {j+1}: - ")))
            b.append(x)
        
        for i in range(m1):
            x = []
            for j in range(n2):
                x.append(0)
            c.append(x)
            
        for i in range(m1):
            for j in range(n2):
                for k in range(m2):
                    c[i][j] += a[i][k] * b[k][i]
                    
        
        print("\nAfter multiplying two matrices: -\n")
        for i in range(m1):
            print(c[i])
        
    except Exception as e:
        print(e)
        
multiplyMatrices()
```

```python
# Question 3

def transopseMatrix():
    try:
        m = int(input("Enter number of rows: - "))
        n = int(input("Enter numbers of columns: - "))
        
        a = []
        b = []
        
        print("\nElements of matrix: - ")
        for i in range(m):
            x = []
            for j in range(n):
                x.append(float(input(f"Enter element {i+1} {j+1}: - ")))
            a.append(x)

            
        for i in range(n):
            x = []
            for j in range(m):
                x.append(a[j][i])
            b.append(x)
            
        print("\nAfter transposing matrix: -\n")
        for i in range(n):
            print(b[i])
        
    except Exception as e:
        print(e)
        
transopseMatrix()
```

```python
# Question 4

def sortWords():
    try:
        m = int(input("Enter number of words: - "))
        
        x = []
        for i in range(m):
            x.append(input("Enter a word: - "))
        
        print("Before sorted: - ", x)
        return sorted(x)
    except Exception as e:
        print(e)

        
print("After sorted: - ", sortWords())
```
